Coleção de arquivos escolhida: Tech

Primeiro descompactamos o arquivo .zip com as coleções de arquivos

```python
import zipfile
import os

# Especifique o nome do arquivo .zip que você fez upload
nome_arquivo_zip = "bbc-fulltext.zip"

# Diretório onde o arquivo .zip está localizado
diretorio_raiz = "/content/"

# Diretório de destino para a extração
diretorio_destino = "/content/extracao/"

# Criar o diretório de destino, se ele não existir
os.makedirs(diretorio_destino, exist_ok=True)

# Caminho completo para o arquivo .zip
caminho_arquivo_zip = os.path.join(diretorio_raiz, nome_arquivo_zip)

# Extrair o conteúdo do arquivo .zip
with zipfile.ZipFile(caminho_arquivo_zip, 'r') as zip_ref:
    zip_ref.extractall(diretorio_destino)

print("Arquivo descompactado com sucesso!")
```

Depois lemos todos os arquivos escolhidos e salvamos seu texto em uma lista

```python
import os

# Diretório onde os arquivos de texto estão localizados
diretorio_arquivos = "/content/extracao/bbc/tech"

# Lista para armazenar o texto de cada arquivo
documentos = []

# Iterar sobre os arquivos na pasta
for nome_arquivo in os.listdir(diretorio_arquivos):
    caminho_completo = os.path.join(diretorio_arquivos, nome_arquivo)

    if os.path.isfile(caminho_completo):
        # Ler o conteúdo do arquivo
        with open(caminho_completo, 'r') as arquivo:
            conteudo = arquivo.read()
            documentos.append(conteudo)  # Adicionar o conteúdo à lista

print("Texto de todos os arquivos foi armazenado na lista.")
```

```python
# Imprimir o número de itens na lista
print("Número de itens na lista de documentos:", len(documentos))
```

```python
# Imports a serem utilizados posteriormente
import nltk
nltk.download('stopwords')
from collections import Counter

from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer
from nltk.tokenize import sent_tokenize , word_tokenize
from nltk.stem import PorterStemmer

STOP_WORDS = set(stopwords.words('english'))

stemmer = PorterStemmer()
nltk.download('punkt')

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Consultas a serem executadas:

- query1: "Best anti-virus"

- query3: "Stolen credit-card number"

Query 1 : "Best anti-virus" ⛳

```python
query = "Best anti-virus"
```

**Primeira rodada: sem retirar stopword e sem "stemming"**

```python
# Inicializar o TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documentos + [query])

# Calcular a similaridade do cosseno entre a consulta e os documentos
cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

# Obter índices ordenados dos documentos por similaridade decrescente
sorted_indices = np.argsort(cosine_similarities[0])[::-1]

numero_resultados = 0
titulos_impressos = set()  # Conjunto para rastrear os títulos dos documentos impressos

# Exibir os documentos ordenados por similaridade
print("Documentos ordenados por similaridade:")
for idx in sorted_indices:
    similarity = cosine_similarities[0][idx]
    if similarity > 0:  # Verificar se a similaridade é maior que zero
        documento = documentos[idx]  # Recupera o documento da lista
        primeira_linha = documento.split('\n', 1)[0]  # Pega a primeira linha do documento

        # Verificar se o título já foi impresso, para evitar duplicados
        if primeira_linha not in titulos_impressos:
            print(f"Similaridade: {similarity:.4f} - Título do artigo: {primeira_linha}")
            titulos_impressos.add(primeira_linha)  # Adicionar o título ao conjunto
            numero_resultados += 1  # Incrementa a contagem

print(f"Total de resultados impressos: {numero_resultados}")
```

É possível observar que 81 resultados foram retornados, com a maior similaridade de 0.3662 e a menor 0.0088.
